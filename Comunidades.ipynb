```python
import networkx as nx
```

```python
lines = []
G = nx.Graph()
with open('../BiBueBa/G.json', 'r') as file:
    lines = file.readlines()

for line_string in lines[1:]:
    line = line_string.split(",")
    G.add_node(line[0])
    G.add_node(line[1])
    print(line[2])
    G.add_edge(line[0], line[1], weight = int(line[2]))

```

```python
print(len(G.edges()))

isolated_nodes = []
cc = nx.connected_components(G)
for connected_component in cc:
    if len(connected_component) <= 3:
        for node in connected_component:
            isolated_nodes.append(node)

for isolated_node in isolated_nodes:
    G.remove_node(isolated_node)
   
print(len(G.edges()))
```

```python
import community
import networkx as nx
import matplotlib.pyplot as plt

#first compute the best partition
partition = community.best_partition(G)


```

```python
count = 0
size = float(len(set(partition.values())))
pos = nx.spring_layout(G)

i = 0
for com in set(partition.values()) :
    if count == 10:
        break
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    nx.draw_networkx_nodes(G, pos, list_nodes, node_size = len(list_nodes),
                                node_color = str(count / size))


nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```
